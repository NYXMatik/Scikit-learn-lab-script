```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Generating sample data
X = 2 * np.random.rand(100, 1)  # Feature (independent variable)
y = 4 + 3 * X + np.random.randn(100, 1)  # Target (dependent variable) with some noise

# Splitting data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Creating and training the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Making predictions
y_pred = model.predict(X_test)

# Plotting the results
plt.scatter(X_test, y_test, color="blue", label="Actual Data")
plt.plot(X_test, y_pred, color="red", linewidth=2, label="Predicted Line")
plt.xlabel("Feature (X)")
plt.ylabel("Target (y)")
plt.legend()
plt.title("Simple Linear Regression with scikit-learn")
plt.show()

# Printing model coefficients
print(f"Intercept: {model.intercept_[0]:.2f}")
print(f"Coefficient: {model.coef_[0][0]:.2f}")
```

```python
R2 = model.score(X_test, y_test)
MAE = np.mean(np.abs(y_test - y_pred))
MSE = ((y_test - y_pred)**2).mean()
RMSE = ((y_test - y_pred)**2).mean()**0.5

print(f"R-squared (R2): {R2:.2f}")
print(f"Mean Absolute Error (MAE): {MAE:.2f}")
print(f"Mean Squared Error (MSE): {MSE:.2f}")
print(f"Root Mean Squared Error (RMSE): {RMSE:.2f}")
```
