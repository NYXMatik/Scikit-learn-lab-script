```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# 1. Generate random data with 3 clusters
data, labels = make_blobs(n_samples=500, centers=10, cluster_std=1.0, random_state=42)

# 2. Plot raw generated blobs (single color)
plt.scatter(data[:, 0], data[:, 1], c='gray', marker='o', edgecolor='k')
plt.title("Generated Blobs (Raw Data)")
plt.show()

# 3. Perform KMeans clustering with 3 clusters
kmeans = KMeans(n_clusters=20, random_state=42) # Here you can choose one of the values [1,2,3,5,7,10,12,15,20]
kmeans.fit(data)
predicted_labels = kmeans.labels_

# 4. Visualize the clustering results
plt.scatter(data[:, 0], data[:, 1], c=predicted_labels, cmap='viridis', marker='o', edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=200, label='Centroids')
plt.legend()
plt.title("KMeans Clustering with make_blobs")
plt.show()
```
